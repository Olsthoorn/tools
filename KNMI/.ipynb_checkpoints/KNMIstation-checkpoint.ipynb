# Using KNMIstation module

The file KNMIstation.py can be run as is. It has some example code near the bottom that is only run when the file is executed on its own.

The module has 3 classes

* KNMI_stations
* KNMI_hourstation
* KNMI_daystation

The **KNMI_stations** class is a derivative of UserDict. When given the file
'./data/KNMI_stations.txt' it reads the KNMI wheather stations file, which has station name, number and coordinates. It can also plot them

```python
import os
os.chdir('/Users/Theo/GRWMODELS/python/KNMI/')
```

```python
#TODO show the stations on the map

import matplotlib.pyplot as plt
import KNMIstation as ws # weather station
from datetime import datetime
```

```python
stationsFile = './data/KNMI_stations.txt'
stations = ws.KNMI_stations(stationsFile)
stations.plot()
plt.show()
```

The **KNMI_hourstation** class will reads a datafile of such a station into a **pandas.DataFrame** that is stored in the instantiated object's attribute **data**


```python
fName_h = './data/uurgeg_240_2001-2010.txt'

stat_h = ws.KNMI_hourstation(fName_h)
stat_h.plot(what='p_air', title='barometer pressure (hourly)')
```

The **KNMI_daystation** class will reads a datafile of such a station into a **pandas.DataFrame** that is stored in the instantiated object's attribute **data**

```python
fName_d = './data/etmgeg_350.txt'

stat_d = ws.KNMI_daystation(fName_d)
stat_d.plot(what='p_air', title='barometer pressure (daily)')
plt.show()
```

Both the **KNMIdaystation** has the columns **'prec'** and **'evap'** that hold the daily precipitation and grass evapotranspiration according to Makkink respectively in mm/d. 

You can use **xlim** to set the start and end times to be plotted. These times need to be **datetimes** from the datetime module.
Use the option **what** to select columns to be plotted

```python
t1 = datetime(1990, 1, 1)
t2 = datetime(2001, 1, 1)
stat_d.plot(what=['prec', 'evap'], xlim=(t1, t2), ylabel='mm/d')
plt.show()
```

To set values or to create a new column for the DataFrame, speciy it directly by accessing teh dataframe in the **data** attribute. Selection can be done directly on the station, no need to access the **data** attribute for that. Here is an example that shows this, by first generating a new column **rch** and then plotting it.

```python
stat_d.data['rch'] = stat_d['prec'] - stat_d['evap']

stat_d.plot(what='rch', xlim=(t1, t2), ylabel='mm/d')
plt.show()
```
