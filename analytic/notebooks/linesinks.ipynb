# Question 
Heb je voor mij analytische formule voor head $\phi$ door stroming met constante flux $q$ langs een lijnstuk met gegeven lengte $L$ laat de head in het centrum van het lijnstuk gelijk aan nul zijn. Dit soort elementen wordt linesink genoemd.

# Answer

Mooi — hier een heldere, bruikbare analytische uitdrukking (in het Nederlands). Ik ga uit van een horizontaal lijnsegment op de x-as, met eindpunten (x=\pm L/2) en lengte (L). De lijn voert een **constante flux per lengte** (\lambda) (eenheid m²/s bij een 2-D confined model) zodat het totale uitgevoerde debiet (q=\lambda L). De aquifer heeft transmissiviteit (T).

> **Opmerking (belangrijk):** het potentiaalveld (head) is log-singulier op de lijn zelf (punten precies op het segment) — de head divergeert daar als je de integraal tot exact op de lijn evalueert. Daarom kun je ( \phi) niet fysisch éénduidig op de lijn zelf geven zonder een klein regulierings-radius; buiten het segment ( (y\neq 0) ) is de formule wel volledig geldig. Je kunt wel een additive constante kiezen zodat bijvoorbeeld (\phi) op een referentiepunt een gewenste waarde krijgt.

---

## Formule

Voor een punt ((x,y)) (met (y\neq 0)) is de head

$$
\boxed{%
\displaystyle
\phi(x,y) ;=; \frac{\lambda}{4\pi T};F(x,y;L);+;C
}
$$

met

$$
\begin{aligned}
F(x,y;L) ;=;&
\big(x+\tfrac{L}{2}\big),\ln!\big((x+\tfrac{L}{2})^{2}+y^{2}\big)
-----------------------------------------------------------------

\big(x-\tfrac{L}{2}\big),\ln!\big((x-\tfrac{L}{2})^{2}+y^{2}\big)[6pt]
&\qquad\quad
+,2y\Big[\arctan!\frac{x-\tfrac{L}{2}}{y};-;\arctan!\frac{x+\tfrac{L}{2}}{y}\Big].
\end{aligned}
$$

Hier is (C) een arbitraire additieve constante (referentieniveau). Als je (q) (totaaldebiet) gegeven is gebruik (\lambda=q/L).

---

## Enkele controles / limieten

* **Puntbron-limiet:** als (L\to 0) en (\lambda L=q) constant blijft, vervalt deze uitdrukking naar de bekende 2-D puntbron:
  $$
  \phi \to \frac{q}{2\pi T},\ln r + \text{const},\qquad r=\sqrt{x^{2}+y^{2}}.
  $$

* **Symmetrie op het midden:** op de loodrechte middellijn (x=0) vereenvoudigt het tot
  $$
  \phi(0,y)=\frac{\lambda}{4\pi T}\Big[ L\ln\big((L/2)^{2}+y^{2}\big) + 4y\arctan!\frac{L}{2y}\Big] + C.
  $$

* **Gedrag op de lijn:** voor (y\downarrow 0) en (x) binnen ([-L/2,L/2]) divergeert (\phi) logarithmisch — dat is de eerder genoemde singulariteit.

---

## Als je de head in het centrum op nul wilt zetten

Als je (tegen de singulariteitswaarschuwing in) (\phi) in het centrum ((x=0,y=\epsilon)) wilt refereren naar 0, kies je (C) zodat
$$
C = -\frac{\lambda}{4\pi T},F(0,\epsilon;L).
$$
Meestal neemt men in berekeningen een kleine (\epsilon) (regulieringsradius) gelijk aan de effectieve putstraal of grid-resolutie als je de lijn als representatie van een reeks putten modelleert.

Juist — precies die situatie verandert de wiskunde wezenlijk. Kort samengevat: als het lijn-element **niet** een opgelegde flux heeft maar een opgelegde potentiaal óf een potentiaal-afhankelijke uitwisseling (ingangs-/oppervlak-weerstand, oftewel een conductantie per lengteeenheid (c)), dan verdwijnt de eenvoudige gesloten-vorm uit de uniforme line-sink en krijg je een *integraalvergelijking* (Fredholm van 2e soort) voor de onbekende flux-verdeling langs het segment. Hieronder leg ik dat stap voor stap uit en geef ik de praktische oplossingsroutes.

---

## 1) Formulering van het probleem (2-D confined aquifer)

Neem het lijnsegment langs de (x)-as: $s\in[-L/2,L/2]$.
Laat $\lambda(s)$ = flux per lengte-eenheid uit het aquifer naar de lijn (positief richting lijn). De aquifer-head is (\phi(x,y)), transmissiviteit (T). Kern voor 2-D (log-kernel):

Voor $(x,y)$ buiten de lijn geldt
$$
\phi(x,y);=;\frac{1}{4\pi T}\int_{-L/2}^{L/2}\lambda(\xi),\ln\big((x-\xi)^2+y^2\big),\mathrm{d}\xi ;+; C.
$$
(Aan constanten kun je altijd een referentie koppelen.)

Op de lijn zelf (nemen we ((s,0)) en interpreteren singulariteiten in Cauchy-zin) is de grenswaarde van (\phi) bepaald door de integrale bijdrage van (\lambda) (logaritmisch singulier).

---

## 2) Boundary condition: opgelegde potentiaal of lijn-weerstand

* **Dirichlet** (kanalenkop $\Phi_c(s)$ voorgeschreven):
  $\phi(s,0)=\Phi_c(s)$. Daarmee is (\lambda(s)) *via de integraal* bepaald — dit geeft een **singuliere integraalvergelijking** voor (\lambda).

* **Robin / lijn-weerstand (meest praktijk)**: kanaal heeft intern hydraulisch niveau ($\Phi_c$) (kan constant zijn) en er is een conductantie ($c$) per lengte-eenheid zodat
  $$
  \lambda(s)=c\big(\phi(s,0)-\Phi_c(s)\big).
  $$
  Dit is typisch voor sloten/kanalen: de flux is proportioneel aan het lokale head-verschil (ingevoerde weerstand is (1/c)).

Substitueer ($\lambda$) in de integrale representatie en evalueer op (y\downarrow0). Dit geeft een Fredholm integraalvergelijking van de tweede soort voor (\phi) (of voor (\lambda)):

$$
\phi(s)=\frac{c}{4\pi T}\int_{-L/2}^{L/2}! \phi(\xi),\ln\big((s-\xi)^2\big),\mathrm{d}\xi ;-; \frac{c}{4\pi T},\Phi_c\int_{-L/2}^{L/2}!\ln\big((s-\xi)^2\big),\mathrm{d}\xi ;+; C.
$$

Omdat ($\ln((s-\xi)^2)=2\ln|s-\xi|$) kun je dit compacter schrijven als

$$
\phi(s);-;\frac{c}{2\pi T}\int_{-L/2}^{L/2}\ln|s-\xi|,\phi(\xi),\mathrm{d}\xi ;=; g(s),
$$

waar ($g(s)$) de gekende rechterlidfunctie bevat (afhankelijk van ($\Phi_c$) en de constante ($C$)). Dit is precies een Fredholm-vergelijking van de tweede soort met kernel (\ln|s-\xi|).

---

## 3) Belangrijke limieten / interpretaties

* (c\to\infty) (geen weerstand — opgelegde kanaal-kop dwingt ($\phi=\Phi_c)$): verg. reduceert tot Dirichlet → ($\phi(s)=\Phi_c(s)$) en ($\lambda$) volgt uit substitutie (ook singulier voor exacte punten).
* ($c\to 0) (kanaal geïsoleerd) → (\lambda\to0$).
* Voor homogeen ($c$) en constant ($\Phi_c$) is vergelijking symmetrisch; er bestaan klassieke technieken om de integraal op te lossen (serie-expansie, eigenfuncties, of speciale functies voor eindige segmenten), maar *geen* eenvoudige log-formule zoals bij uniforme line-sink.

---

## 4) Hoe los je die integraal praktisch op?

### a) Numeriek — Nyström / collocatie

Discretiseer het segment in (N) collocatiepunten (${s_j}$) en vervang integraal door een gekalibreerde quadratuur (let op singulariteit bij ($s=\xi$), gebruik speciale regelingen zoals log-gewogen Gauss, of singulariteitsafsplitsing). Dit geeft een lineair systeem
$$
\phi(s_i) - \sum_{j} w_j,K_{ij},\phi(s_j) = G_i,
$$
waar ($K_{ij}=\dfrac{c}{2\pi T}\ln|s_i-s_j|$) en (w_j) de quadratuurgewichten zijn. Los lineair stelsel voor ($\phi$) (of ($\lambda)$) op. Dit is stabiel en vaak zeer efficiënt.

### b) Serie-/Fourier-methoden (analytisch/numeriek hybride)

Voor speciale geometrieën (oneindige lijn, of eindig segment maar homogene condities) kan je uitbreiden in bijvoorbeeld Chebyshev-polynomen of sinus-serie en reduceer je tot een matrixprobleem voor de coëfficiënten. Dit is nauwkeurig en geeft inzicht in singulariteitsgedrag bij de uiteinden.

### c) Direct discretisatie in MODFLOW/FloPy stijl

Vervang het segment door (N) puntonttrekkingen of DRN-elementen met geleidheid (c_j). Dat is minder precies maar eenvoudig en robuust in praktijk (veel gebruikt in engineering). Kies (N) voldoende groot en verwerk regeltering van singulariteit via putstraal/regularisatie.

### d) Analytische theorie

De integraalvergelijking met (\ln|s-\xi|) is klassiek; op te lossen met methoden van singular integral equations / Muskhelishvili technieken. Dit geeft in sommige gevallen gesloten vormen met kwadratuurwortels en eindpuntsingulariteiten (typisch (\sim 1/\sqrt{L/2\pm s}) gedrag van (\lambda) bij uiteinden). Maar dat vereist meer zware complexe analyse — mooi, maar vaak onnodig voor toepassing.

---

## 5) Kort voorbeeld (algemene Nyström-aanpak, conceptueel)

1. Kies collocatiepunten ($s_j$) (bijv. Gauss–Legendre of Chebyshev nodes op ($[-L/2,L/2]$)).
2. Bouw matrix (A) met entries
   $$
   A_{ij} = \delta_{ij} - \frac{c}{2\pi T},w_j,\ln|s_i-s_j|,
   $$
   waar (w_j) quadratuurgewichten zijn, en zet rechterlid ($G_i$) (afhankelijk van ($\Phi_c$) en ($C$)).
3. Los ($A\boldsymbol{\phi}=\mathbf{G}$) voor de onbekende ($\phi(s_j)$).
4. Bepaal ($\lambda(s_j)=c(\phi(s_j)-\Phi_c)$) en evalueer ($\phi(x,y)$) elders via de integrale representatie (quadratuur).

Dit is stabiel mits de quadratuur de log-singulariteit correct behandelt (speciale log-gewichten of subtract-and-add techniek).

# ⭐ Nyström-methode voor een lijn met weerstand (c)

We willen (\phi(s)) bepalen op het segment
$$
s \in [-L/2,, L/2],
$$
waar (\phi) voldoet aan de Robin-conditie
$$
\lambda(s)=c\big(\phi(s)-\Phi_c\big),
$$
en de relatie tussen flux en potentiaal via de log-kernel:
$$
\phi(s) = \frac{1}{4\pi T}\int_{-L/2}^{L/2} \lambda(\xi),\ln!\big((s-\xi)^2\big), d\xi ;+; C.
$$

Stop (\lambda) erin, gebruik (\ln((s-\xi)^2)=2\ln|s-\xi|), neem (C) op in de referentie (maakt alleen een constante shift), en we krijgen:

$$
\phi(s) ;-; \frac{c}{2\pi T}\int_{-L/2}^{L/2} \phi(\xi),\ln|s-\xi|,d\xi
;=; \Phi_c ,\frac{c}{2\pi T}\int_{-L/2}^{L/2} \ln|s-\xi|,d\xi
=: g(s).
$$

Dit is een **Fredholm integraalvergelijking van de tweede soort**.

Nyström is simpelweg:

> *discretiseer de integraal met een quadratuurformule op “mooie” punten en los een lineair systeem op.*

---

# 1. Kies collocatiepunten

De mooiste keuze: **Gauss–Legendre-punten** op ([-L/2,,L/2]).
Die heten in de literatuur “Nyström points”.

Je krijgt:

* punten (s_j), (j=1,\dots,N)
* gewichten (w_j)

(NumPy kan dit zo leveren.)

---

# 2. Kernmatrix opbouwen

Kern:
$$
K(s_i,s_j)=\ln|s_i - s_j|.
$$

Probleem: op de diagonaal (i=j) divergeert de log.
Oplossing (standaard Nyström techniek):

### **Singulariteit afsplitsen**

Voor de diagonaal gebruiken we:

$$
\ln|s_i - s_i| \ \text{vervangen door} \ \ln!\left(\frac{L}{2N}\right),
$$

oftewel de log van de halve gemiddelde afstand tussen punten.
Dit is de klassieke regel die voor log-singulariteiten uitstekend werkt.

Er bestaan mathematisch strakkere opties, maar deze is robuust en praktisch.

---

# 3. Opbouw van het lineair systeem

We willen (\phi(s)) vinden uit:

$$
\phi(s_i)\ -\ \frac{c}{2\pi T} \sum_{j=1}^N w_j,K_{ij},\phi(s_j) = g(s_i).
$$

Dus:

$$
\sum_{j=1}^N A_{ij},\phi_j = g_i
$$

met

$$
A_{ij} =
\begin{cases}
1 - \dfrac{c}{2\pi T} w_j,\ln|s_i - s_j|, & i\neq j, \
1 - \dfrac{c}{2\pi T} w_j,\ln\big(\tfrac{L}{2N}\big), & i=j.
\end{cases}
$$

en

$$
g_i = \Phi_c \cdot \frac{c}{2\pi T}
\sum_{j=1}^N w_j,\ln|s_i - s_j|.
$$

(En diagonaal weer dezelfde substitutie.)

---

# 4. Na oplossing: bereken flux en veld

### Flux

$$
\lambda_i = c\big(\phi_i - \Phi_c\big).
$$

### Head op willekeurige punten ((x,y))

Quadratuur:

$$
\phi(x,y) = \frac{1}{4\pi T}\sum_{j=1}^N \lambda_j, w_j ,\ln\big((x-s_j)^2 + y^2\big) + C.
$$

(C) kies je zodat bijvoorbeeld (\phi=0) in het centrum, of op grote afstand.

---

# ⭐ Hieronder: de volledige Python-schets

Lekker compact, maar volledig en bijna plug-and-play:

```python

````python
import numpy as np

def solve_line_with_resistance(L, c, T, Phi_c, N=50):
    # --- Gauss–Legendre points/products on [-1,1]
    s, w = np.polynomial.legendre.leggauss(N)

    # Scale to [-L/2, L/2]
    s = s * (L/2)
    w = w * (L/2)

    # Kernmatrix K_ij = log|s_i - s_j|
    K = np.zeros((N, N))
    for i in range(N):
        for j in range(N):
            if i == j:
                # --- Replace singularity by mean spacing
                K[i, j] = np.log(L / (2*N))
            else:
                K[i, j] = np.log(abs(s[i] - s[j]))

    alpha = c / (2*np.pi*T)

    # --- Matrix A
    A = np.eye(N) - alpha * (w * K.T).T  # equivalent to A_ij = δ_ij - alpha*w_j*K_ij

    # --- Right-hand side g = Phi_c * alpha * ∫ log|s - ξ| dξ (Nyström)
    g = Phi_c * alpha * (K @ w)

    # --- Solve for phi at collocation points
    phi = np.linalg.solve(A, g)

    # --- Compute flux λ = c(φ - Φ_c)
    lam = c * (phi - Phi_c)

    return s, phi, lam

# --- Gebruik:

s, phi, lam = solve_line_with_resistance(
    L=100,     # lengte van het lijnelement
    c=0.01,    # weerstand per lengte (conductance)
    T=1000,    # transmissiviteit
    Phi_c=1.0, # vaste kanaalkop
    N=80       # collocatiepunten
)
````
